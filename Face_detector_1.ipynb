```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('your image name')
```

```python
plt.imshow(img)
```

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray,cmap='gray')
```

```python
haar=cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```python
face=haar.detectMultiScale(img_gray,1.1,5)
```

```python
#get number of detected faces in the picture
print(len(face))
```

```python
for (x,y,w,h) in face:
    img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
img_temp=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_temp)    
```
